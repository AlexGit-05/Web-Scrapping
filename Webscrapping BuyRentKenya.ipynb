```python
# loading libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# specify url
url = "https://www.buyrentkenya.com/property-for-rent"

# get request to URL
page = requests.get(url)

# Html layout
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
# Define an empty list to store dictionaries for each listing
listing_data = []

# Extract all <div> tags with class "listing-card mb-5 w-full md:mx-0"
listing_cards = soup.find_all('div', class_="listing-card mb-5 w-full md:mx-0")

# Loop through indices 1 to 17
for listing_cards in listing_cards:
    # Extract all <a> tags with class "no-underline"
    no_underline = listing_cards.find_all('a', class_="no-underline")

    # Extract Title
    title = no_underline[0].text.strip()

    # Extract Price
    price = no_underline[2].text.replace('\n', '')

    # Extract features
    feature = no_underline[4].text.strip()

    # Extract realestate organization
    organization = no_underline[0]['wire:click.prevent'].split('\\')[11].replace('u0022', '')

    # Extract location of house
    location = no_underline[0]['wire:click.prevent'].split('\\')[89].replace('u0022', '')

    # Store the data in a dictionary
    listing_info = {
        'Title': title,
        'Organization': organization,
        'Location': location,
        'Feature': feature,
        'Price': price,
    }

    # Append the dictionary to the list
    listing_data.append(listing_info)

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(listing_data)

# DataFrame
df
```

```python
# Define an empty list to store dictionaries for each listing
all_listing_data = []

# Loop through pages 2 to 399
for page_number in range(2, 399):
    # Construct the URL for the current page
    current_url = f"https://www.buyrentkenya.com/property-for-rent?page={page_number}"

    # Get request to the current URL
    page = requests.get(current_url)

    # Check if the request was successful (status code 200)
    if page.status_code == 200:
        soup = BeautifulSoup(page.text, 'html.parser')

        # Extract all <div> tags with class "listing-card mb-5 w-full md:mx-0"
        listing_cards = soup.find_all('div', class_="listing-card mb-5 w-full md:mx-0")

        # Loop through listing cards on the current page
        for listing_card in listing_cards:
            # Extract all <a> tags with class "no-underline"
            no_underline = listing_card.find_all('a', class_="no-underline")

            # Extract Title
            title = no_underline[0].text.strip()

            # Extract Price
            price = no_underline[2].text.replace('\n', '')

            # Extract features
            feature = no_underline[4].text.strip()

            # Extract realestate organization
            organization = no_underline[0]['wire:click.prevent'].split('\\')[11].replace('u0022', '')

            # Extract location of house
            location = no_underline[0]['wire:click.prevent'].split('\\')[89].replace('u0022', '')

            # Store the data in a dictionary
            listing_info = {
                'Title': title,
                'Organization': organization,
                'Location': location,
                'Feature': feature,
                'Price': price
            }

            # Append the dictionary to the list
            all_listing_data.append(listing_info)

    else:
        print(f"Failed to retrieve page {page_number}. Status code: {page.status_code}")

# Create a DataFrame from the list of dictionaries
df1 = pd.DataFrame(all_listing_data)

# DataFrame
df1
```

```python
# Row bind
full_dataset= pd.concat([df, df1], ignore_index=True)
full_dataset.to_csv('Property for rent in Kenya.csv', index=False)
```
